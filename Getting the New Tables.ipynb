```python
import pypyodbc as od
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
DRIVER_NAME = 'SQL SERVER'
SERVER_NAME = 'ADNAN\SQLEXPRESS'
DATABASE_NAME = 'FitBit'

connection_string = f"""
DRIVER={{{DRIVER_NAME}}};
SERVER={SERVER_NAME};
DATABASE={DATABASE_NAME};
Trust_Connection=yes;
"""

conn = od.connect(connection_string)
print(conn)
```

```python
cursor = conn.cursor()
```

```python
df1 = pd.read_sql('SELECT * FROM [Daily Activity Info]', conn)
df1.head()
```

```python
df2 = pd.read_sql('SELECT * FROM [Hourly Activity Info]', conn)
df2.head()
```

```python
df1.to_csv('Daily Activity Info.csv', index=False)
```

```python
df2.to_csv('Hourly Activity Info.csv', index=False)
```
