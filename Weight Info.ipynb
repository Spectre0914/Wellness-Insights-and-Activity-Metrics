```python
#Converting the column to datetime according the SQL Standard
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
df.dtypes
```

```python
#Reducing the Float values to only 2 Decimal Places to avoid clutter
df['WeightKg'] = df['WeightKg'].round(2)
df['WeightPounds'] = df['WeightPounds'].round(2)
df['BMI'] = df['BMI'].round(2)
```

```python
df['Fat'].isnull().mean() * 100
```

```python
df.drop('Fat', axis = 1, inplace = True)
```

```python
df.drop('LogId', axis = 1, inplace = True)
```
